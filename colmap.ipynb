```python
import torch
import torchvision.transforms as T
from torchvision.transforms.functional import InterpolationMode
from PIL import Image
import numpy as np
import os
import glob
from tqdm.auto import tqdm # Progress bar
import torch.nn as nn
MODEL_VARIANT = 'dinov2_vitb14' # A good balance

IMAGE_DIR = '/kaggle/input/image-matching-challenge-2025/train/ETs' # <--- CHANGE THIS to your image folder path

IMAGE_EXT = '.png'

FEATURES_OUTPUT_FILE = 'dinov2_extracted_features.npz' # Save as compressed numpy archive

if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
    device = torch.device("cpu")
    print("Using CPU")

print(f"Loading DINOv2 model: {MODEL_VARIANT}...")
try:
    dinov2_model, dinov2_transforms = torch.hub.load('facebookresearch/dinov2', MODEL_VARIANT, pretrained=True)
    if dinov2_transforms is None:
      raise ValueError("Transforms were not loaded automatically.") # Force fallback

except Exception as e:
     print(f"Warning: Could not load model/transforms automatically ({e}). Attempting manual transform creation.")
     dinov2_model = torch.hub.load('facebookresearch/dinov2', MODEL_VARIANT, pretrained=True)
     dinov2_transforms = T.Compose([
            T.Resize(256, interpolation=InterpolationMode.BICUBIC),
            T.CenterCrop(224), 
            T.ToTensor(),
            T.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)), 
        ])
     print("Using manually defined transforms (Resize 256 -> CenterCrop 224 -> ToTensor -> Normalize).")


dinov2_model.eval() 
dinov2_model = nn.DataParallel(dinov2_model)
dinov2_model.to(device)
print("Model and transforms loaded successfully.")

def preprocess_image(image_path, transforms):
    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = transforms(img).unsqueeze(0) 
        return img_tensor
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None

@torch.no_grad() 
def extract_features(model, image_tensor, target_device):
    """Extracts features (CLS token embedding) from a preprocessed image tensor."""
    image_tensor = image_tensor.to(target_device)
    features = model(image_tensor) # Forward pass

    if isinstance(features, dict):
        if 'x_norm_clstoken' in features:
             image_features = features['x_norm_clstoken']
        elif 'x_prenorm' in features: 
             image_features = features['x_prenorm'][:, 0] 
        else:
             keys = list(features.keys())
             print(f"Warning: Feature dict keys: {keys}. Trying to find CLS token.")
             potential_feature_key = keys[0] 
             print(f"Assuming features are in key: '{potential_feature_key}'")
             feature_output = features[potential_feature_key]
             if feature_output.ndim == 3 and feature_output.shape[1] > 0:
                  image_features = feature_output[:, 0] 
             elif feature_output.ndim == 2:
                  image_features = feature_output 
             else:
                  raise ValueError(f"Unexpected feature tensor shape: {feature_output.shape}")

    elif isinstance(features, torch.Tensor):
        if features.ndim == 3:
            image_features = features[:, 0]
        elif features.ndim == 2:
             image_features = features 
        else:
            raise ValueError(f"Unexpected feature tensor shape: {features.shape}")
    else:
        raise TypeError(f"Unexpected output type from model: {type(features)}")

    return image_features.squeeze(0).cpu().numpy()

image_paths = sorted(glob.glob(os.path.join(IMAGE_DIR, f'*{IMAGE_EXT}')))
if not image_paths:
    print(f"Error: No images found in {IMAGE_DIR} with extension {IMAGE_EXT}")
    exit()

print(f"Found {len(image_paths)} images to process.")

extracted_features_list = []
corresponding_image_ids = []
failed_images = []

for img_path in tqdm(image_paths, desc="Extracting Features"):
    image_tensor = preprocess_image(img_path, dinov2_transforms)
    if image_tensor is not None:
        try:
            features_np = extract_features(dinov2_model, image_tensor, device)
            extracted_features_list.append(features_np)
            image_id = os.path.basename(img_path)
            corresponding_image_ids.append(image_id)
        except Exception as e:
            print(f"Error extracting features for {img_path}: {e}")
            failed_images.append(img_path)
    else:
        failed_images.append(img_path) 

print(f"\nSuccessfully processed {len(extracted_features_list)} images.")
if failed_images:
    print(f"Failed to process {len(failed_images)} images:")

if extracted_features_list:
    all_features_np = np.vstack(extracted_features_list)

    print(f"\nSaving {all_features_np.shape[0]} features (Shape: {all_features_np.shape}) and {len(corresponding_image_ids)} image IDs to {FEATURES_OUTPUT_FILE}...")

    np.savez_compressed(
        FEATURES_OUTPUT_FILE,
        features=all_features_np,
        image_ids=np.array(corresponding_image_ids) # Save IDs as numpy array too
    )
    print("Features and IDs saved successfully.")

    print("\nTo load the data later:")
    print(f"data = np.load('{FEATURES_OUTPUT_FILE}', allow_pickle=True)")
    print("features_array = data['features']")
    print("image_id_array = data['image_ids']")

else:
    print("\nNo features were extracted, nothing to save.")
print("\nFeature extraction complete.")
```

```python
import numpy as np
import faiss
import time
import os

NPZ_FILE_PATH = '/kaggle/working/dinov2_extracted_features.npz' 
EMBEDDING_KEY = 'features'
METADATA_KEY = 'filenames'  

N_NEIGHBORS = 5  
INDEX_TYPE = 'IndexFlatL2'
INDEX_SAVE_PATH = 'my_faiss_index.index' 

print(f"Loading data from: {NPZ_FILE_PATH}")
try:
    data = np.load(NPZ_FILE_PATH)
    embeddings = data['features']
    print(f"Loaded embeddings with shape: {embeddings.shape}")

    metadata = None
    if 'image_ids' and 'image_ids' in data:
        metadata = data['image_ids']
        if len(metadata) != len(embeddings):
            print(f"Warning: Metadata count ({len(metadata)}) does not match embedding count ({len(embeddings)}).")
        else:
            print(f"Loaded metadata for {len(metadata)} items.")

except FileNotFoundError:
    print(f"Error: NPZ file not found at {NPZ_FILE_PATH}")
    exit()
except KeyError:
    print(f"Error: Key '{EMBEDDING_KEY}' not found in the NPZ file.")
    print(f"Available keys: {list(data.keys())}")
    exit()

if embeddings.dtype != np.float32:
    print("Converting embeddings to float32...")
    embeddings = embeddings.astype(np.float32)

if not embeddings.flags['C_CONTIGUOUS']:
    print("Making embeddings C-contiguous...")
    embeddings = np.ascontiguousarray(embeddings)

num_embeddings, dimension = embeddings.shape
print(f"Number of embeddings: {num_embeddings}, Dimension: {dimension}")

if os.path.exists(INDEX_SAVE_PATH):
    print(f"Loading existing Faiss index from: {INDEX_SAVE_PATH}")
    index = faiss.read_index(INDEX_SAVE_PATH)
    if index.d != dimension:
         print(f"Error: Loaded index dimension ({index.d}) does not match data dimension ({dimension}). Rebuilding index.")
         index = None 
    elif not index.is_trained:
         print("Warning: Loaded index is not trained. Training required before adding (should not happen for saved trained indexes). Rebuilding.")
         index = None 
    elif index.ntotal != num_embeddings and INDEX_TYPE != 'IndexHNSWFlat': # HNSW can be added to incrementally easier
         print(f"Warning: Number of vectors in loaded index ({index.ntotal}) differs from data ({num_embeddings}). Rebuilding index.")
         index = None 

if 'index' not in locals() or index is None:
    print(f"\nBuilding Faiss index of type: {INDEX_TYPE}")
    start_time = time.time()

    if INDEX_TYPE == 'IndexFlatL2':
        index = faiss.IndexFlatL2(dimension)

    else:
        print(f"Error: Unknown INDEX_TYPE '{INDEX_TYPE}'")
        exit()

    print(f"Adding {num_embeddings} embeddings to the index...")
    index.add(embeddings)
    end_time = time.time()

    print(f"Index built and populated successfully.")
    print(f" - Index type: {INDEX_TYPE}")
    print(f" - Total vectors in index: {index.ntotal}")
    print(f" - Index is trained: {index.is_trained}")
    print(f" - Build time: {end_time - start_time:.2f} seconds")

    print(f"Saving index to: {INDEX_SAVE_PATH}")
    faiss.write_index(index, INDEX_SAVE_PATH)

print(f"\nPerforming ANN search for {N_NEIGHBORS} neighbors...")

num_queries = max(3, num_embeddings)
query_vectors = embeddings[:num_queries]

start_time = time.time()
distances, indices = index.search(query_vectors, N_NEIGHBORS)
print(indices)
end_time = time.time()

print(f"Search completed in {end_time - start_time:.4f} seconds for {num_queries} queries.")
li = []

print("\nSearch Results:")
for i in range(num_queries):
    print(f"\nQuery {i} (vector index {i}):") 
    if metadata is not None:
         try:
             print(f"  Query Metadata: {metadata[i]}")
         except IndexError:
              print(f"  Query Metadata: N/A (index {i} out of bounds for metadata)")

    print(f"  Nearest Neighbors (Index, Distance):")
    for j in range(N_NEIGHBORS):
        neighbor_idx = indices[i, j]
        distance = distances[i, j]

        if neighbor_idx == i:
            print(f"    - Self (Index {neighbor_idx}), Distance: {distance:.4f}")
            continue

        neighbor_info = f"    - Neighbor {j+1}: Index {neighbor_idx}, Distance: {distance:.4f}"
        if metadata is not None:
            try:
                neighbor_info += f", Metadata: {metadata[neighbor_idx]}"
            except IndexError:
                neighbor_info += f", Metadata: N/A (index {neighbor_idx} out of bounds for metadata)"
        print(neighbor_info)
        li.append([metadata[i],metadata[neighbor_idx]])
    if i == 0:
         print(f"\nRaw distances for query 0: {distances[i]}")
         print(f"Raw indices for query 0:   {indices[i]}")
print(li)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_matches(image0, image1, points0, points1):
    # Convert tensors to numpy arrays and scale image for Matplotlib (from [0,1] to [0,255])
    image0_np = (image0.cpu().numpy().transpose(1, 2, 0) * 255).astype(np.uint8)
    image1_np = (image1.cpu().numpy().transpose(1, 2, 0) * 255).astype(np.uint8)

    # Create a combined image to display matches on one panel
    combined_image = np.concatenate([image0_np, image1_np], axis=1)
    plt.figure(figsize=(12, 8))
    plt.imshow(combined_image)

    # Plot the matches using lines
    for (pt0, pt1) in zip(points0.cpu().numpy(), points1.cpu().numpy()):
        x0, y0 = pt0
        x1, y1 = pt1
        plt.plot([x0, x1 + image0_np.shape[1]], [y0, y1], 'r', linewidth=0.5)

    plt.axis('off')
    plt.title("Matches", fontsize=16)
    plt.show()

plot_matches(image0, image1, points0, points1)
```

```python
import networkx as nx
import networkx.algorithms.community as nx_comm 
import matplotlib.pyplot as plt 

if G is None or G.number_of_nodes() == 0:
    print("Graph object 'G' does not exist or is empty. Cannot perform clustering.")
    final_clusters = []
elif G.number_of_edges() == 0:
    print("Graph has nodes but no edges. No communities can be formed. Each node is isolated.")
    final_clusters = []
else:
    print("\nApplying community detection (Louvain algorithm)...")
    print("Using edge weights (inlier counts) for community structure.")

    communities = nx_comm.louvain_communities(G, weight='weight', resolution=0.3, seed=42)

    communities = sorted(communities, key=len, reverse=True)

    print(f"Found {len(communities)} raw communities.")

    MIN_CLUSTER_SIZE = 3  
    print(f"Filtering out clusters with fewer than {MIN_CLUSTER_SIZE} images...")

    final_clusters = []
    original_cluster_indices = [] 
    for i, community_set in enumerate(communities):
        if len(community_set) >= MIN_CLUSTER_SIZE:
            final_clusters.append(list(community_set))
            original_cluster_indices.append(i) # Store the index from the 'communities' list

    print(f"\nRetained {len(final_clusters)} clusters meeting the size requirement (>= {MIN_CLUSTER_SIZE} images).")

    print("\nFinal Image Clusters for Reconstruction:")
    if not final_clusters:
        print("  No clusters found meeting the minimum size requirement.")
    else:
        for i, cluster in enumerate(final_clusters):
            cluster.sort() # Sort the list of image names in-place
            print(f"  Cluster {i+1} (Original Index: {original_cluster_indices[i]}, Size: {len(cluster)}):")
            # Print first few and last few images if cluster is large
            if len(cluster) > 10:
                 print(f"    Images: {cluster[:5]} ... {cluster[-5:]}")
            else:
                 print(f"    Images: {cluster}")

    if G.number_of_nodes() > 0 and G.number_of_nodes() < 150 and final_clusters:
        print("\nVisualizing graph with final clusters...")
        plt.figure(figsize=(15, 15))
        pos = nx.spring_layout(G, k=0.6, iterations=50, seed=42) # Use same seed for layout consistency

        node_colors = {}
        color_map = plt.get_cmap('viridis', len(final_clusters) + 1)

        for cluster_id, cluster_nodes in enumerate(final_clusters):
            for node in cluster_nodes:
                node_colors[node] = color_map(cluster_id)

        nodes_in_final_clusters = set().union(*final_clusters)
        default_color = (0.8, 0.8, 0.8, 0.5) # Light gray RGBA
        node_color_list = [node_colors.get(node, default_color) for node in G.nodes()]

        nx.draw_networkx_nodes(G, pos, node_size=50, node_color=node_color_list)

        weights = [G[u][v]['weight'] for u,v in G.edges()]
        max_weight = max(weights) if weights else 1
        normalized_weights = [(w / max_weight * 2) + 0.1 for w in weights] # Thin lines
        nx.draw_networkx_edges(G, pos, width=normalized_weights, edge_color='gray', alpha=0.3)

        plt.title(f"Image Graph with {len(final_clusters)} Filtered Communities (Min Size {MIN_CLUSTER_SIZE})")
        plt.axis('off')
        plt.show()
    elif G.number_of_nodes() >= 150:
         print("\nSkipping cluster visualization (graph too large).")
```

```python
import pycolmap
import os
import numpy as np
from PIL import Image
import itertools
import sqlite3 


IMAGE_DIR = '/kaggle/input/image-matching-challenge-2025/train/ETs' 
WORKSPACE_DIR = '/kaggle/working/' 
CLUSTER_INDEX_TO_PROCESS = 0 

ASSUME_SINGLE_CAMERA = True
CAMERA_MODEL = 'SIMPLE_RADIAL'
ESTIMATED_FOCAL_LENGTH_FACTOR = 1.2 
PRINCIPAL_POINT_IS_CENTER = True


def get_image_path(name):
    return os.path.join(IMAGE_DIR, name)


import pycolmap
import os
import numpy as np
from PIL import Image
import itertools
import sqlite3 

def prepare_colmap_input_for_cluster(cluster, cluster_id, workspace_base, image_base_dir,
                                     features_dict, matches_dict):
    """Prepares COLMAP database and workspace for a single image cluster."""

    print(f"\n--- Preparing COLMAP input for Cluster {cluster_id} ---")
    print(f"Number of images in cluster: {len(cluster)}")

    cluster_workspace = os.path.join(workspace_base, f"cluster_{cluster_id:03d}")
    os.makedirs(cluster_workspace, exist_ok=True)

    database_path = os.path.join(cluster_workspace, "database.db")
    if os.path.exists(database_path):
        print(f"Database already exists at {database_path}. Removing.")
        os.remove(database_path)

    db = pycolmap.Database(database_path)

    print("Preparing cameras...")
    name_to_image_id = {}
    camera_map = {} 

    if ASSUME_SINGLE_CAMERA:
        first_image_name = cluster[0]
        print("Is is this?:",first_image_name)
        first_image_path = os.path.join(image_base_dir, first_image_name)
        try:
            with Image.open(first_image_path) as img:
                W, H = img.size
        except FileNotFoundError:
             print(f"Error: Cannot open image {first_image_path} to get dimensions.")
             return None
        except Exception as e:
            print(f"Error reading image {first_image_path}: {e}")
            return None

        focal_length = ESTIMATED_FOCAL_LENGTH_FACTOR * max(W, H)
        cx = W / 2.0 if PRINCIPAL_POINT_IS_CENTER else W / 2.0 
        cy = H / 2.0 if PRINCIPAL_POINT_IS_CENTER else H / 2.0 
        params = [focal_length, cx, cy,0.0] 

        cam = pycolmap.Camera(model=CAMERA_MODEL, width=W, height=H, params=params)

        
        single_camera_id = db.write_camera(cam)

        for img_name in cluster:
            camera_map[img_name] = single_camera_id
        print(f"Defined single camera model '{CAMERA_MODEL}' (Assigned ID: {single_camera_id}) with estimated params: W={W}, H={H}, f={params[0]:.2f}, cx={params[1]:.2f}, cy={params[2]:.2f}")

    else:
        print("Error: Per-image camera parameters from EXIF not implemented yet.")
        return None
    print("Adding images and keypoints to database...")
    num_images_added = 0
    for image_name in cluster: 

        if image_name not in features_dict:
            print(f"Warning: Features not found for image {image_name}. Skipping.")
            continue

        if image_name not in camera_map:
             print(f"Warning: Camera ID not found for image {image_name}. Skipping.")
             continue 

        keypoints = features_dict[image_name].astype(np.float32) # N x 2

        if keypoints.shape[1] != 2 or keypoints.ndim != 2:
             print(f"Warning: Invalid keypoint shape {keypoints.shape} for {image_name}. Expected (N, 2). Skipping.")
             continue  

        camera_id = camera_map[image_name]
        img_obj = pycolmap.Image(name=image_name, camera_id=camera_id)

        image_id = db.write_image(img_obj)
        name_to_image_id[image_name] = image_id

        db.write_keypoints(image_id, keypoints)
        num_images_added += 1
    print("Adding verified matches (two-view geometry) to database...")
    num_added_matches = 0
    
    for name1, name2 in itertools.combinations(cluster, 2):
        if name1 not in name_to_image_id or name2 not in name_to_image_id:
            continue
    
        pair_key = tuple(sorted((name1, name2)))
        if pair_key in matches_dict:
            inlier_matches = matches_dict[pair_key].astype(np.uint32) # K x 2 array of indices
            if inlier_matches.shape[1] != 2 or inlier_matches.ndim != 2:
                 print(f"Warning: Invalid matches shape {inlier_matches.shape} for pair {pair_key}. Skipping.")
                 continue
            if len(inlier_matches) == 0:
                 continue
    
            image_id1 = name_to_image_id[name1]
            image_id2 = name_to_image_id[name2]
    
            db.write_matches(image_id1, image_id2, inlier_matches)
    
            geom = pycolmap.TwoViewGeometry() 
            geom.inlier_matches = inlier_matches  
    
            geom.config = 2

            db.write_two_view_geometry(image_id1, image_id2, geom)
    
            num_added_matches += 1
    
    print(f"Written verified matches for {num_added_matches} pairs.")
    if num_added_matches == 0:
         print("Warning: No verified matches were written to the database for this cluster!")

    # Close the database connection
    db.close()
    print(f"\nSuccessfully wrote COLMAP database to: {database_path}")
    return database_path, cluster_workspace


if not final_clusters:
    print("No clusters found to process.")
else:
    current_cluster_list = final_clusters[CLUSTER_INDEX_TO_PROCESS]

    if 'all_features' not in locals() or 'all_verified_matches' not in locals():
         print("Error: 'all_features' or 'all_verified_matches' not found.")
         print("Please ensure these dictionaries are populated from the feature extraction and matching steps.")
    elif not all_features or not all_verified_matches:
         print("Warning: 'all_features' or 'all_verified_matches' dictionary is empty.")
    else:
        result = prepare_colmap_input_for_cluster(
            current_cluster_list,
            CLUSTER_INDEX_TO_PROCESS,
            WORKSPACE_DIR,
            IMAGE_DIR,
            all_features,
            all_verified_matches
        )

        if result:
            database_path, cluster_workspace = result
            print(f"\nInput preparation complete for cluster {CLUSTER_INDEX_TO_PROCESS}.")
            print(f"Workspace: {cluster_workspace}")
            print(f"Database: {database_path}")
            print("\nYou can now run COLMAP steps (e.g., mapper) using this database.")
        else:
            print(f"\nInput preparation failed for cluster {CLUSTER_INDEX_TO_PROCESS}.")
```

```python
import pycolmap
import os
import numpy as np
from PIL import Image
import itertools
import sqlite3
import time
import traceback
from scipy.spatial.transform import Rotation as R_sci 
import numpy as np
from copy import deepcopy

def qvec2rotmat(q):
    """(w,x,y,z) → 3×3 rotation matrix (no SciPy needed)."""
    w, x, y, z = q
    return np.array(
        [[1-2*(y*y+z*z), 2*(x*y-z*w),     2*(x*z+y*w)],
         [2*(x*y+z*w),   1-2*(x*x+z*z),   2*(y*z-x*w)],
         [2*(x*z-y*w),   2*(y*z+x*w),     1-2*(x*x+y*y)]],
        dtype=np.float64)

def get_pose_from_image(img):
    if hasattr(img, "qvec") and hasattr(img, "tvec"):
        qvec = np.asarray(img.qvec, dtype=np.float64)
        tvec = np.asarray(img.tvec, dtype=np.float64)
        return qvec2rotmat(qvec), tvec

    if hasattr(img, "cam_from_world"):
        print("It has cam from world")
    if all(hasattr(img, k) for k in ("qw","qx","qy","qz",
                                 "tx", "ty", "tz")):
        qvec = np.array([img.qw, img.qx, img.qy, img.qz], dtype=np.float64)
        tvec = np.array([img.tx, img.ty, img.tz], dtype=np.float64)
        return qvec2rotmat(qvec), tvec
    if all(hasattr(img, k) for k in ("qw","qx","qy","qz",
                                     "tx","ty","tz")):
        qvec = np.array([getattr(img, k) for k in ("qw","qx","qy","qz")],
                        dtype=np.float64)
        tvec = np.array([getattr(img, k) for k in ("tx","ty","tz")],
                        dtype=np.float64)
        return qvec2rotmat(qvec), tvec

    if hasattr(img, "T"):              
        T = np.asarray(img.T, dtype=np.float64)
        R = T[:3, :3]
        t = T[:3, 3]
        return R, t

    if hasattr(img, "cam_from_world"):
        T_cw = np.asarray(img.cam_from_world.matrix(), dtype=np.float64)
        R_cw, t_cw = T_cw[:3, :3], T_cw[:3, 3]
        return R_cw, t_cw

    return None        


def image_pose(img):
    qvec = np.array([img.qw, img.qx, img.qy, img.qz], dtype=np.float64)
    tvec = np.array([img.tx, img.ty, img.tz], dtype=np.float64)
    R    = qvec2rotmat(qvec)         
    return R, tvec

```

```python
from copy import deepcopy
def run_sfm_and_extract_poses(database_path, image_dir, cluster_workspace):
    print(f"\n--- Running SfM for cluster in workspace: {cluster_workspace} ---")
    print(f"Using database: {database_path}")
    print(f"Image directory: {image_dir}")

    sparse_model_path = os.path.join(cluster_workspace, "sparse")
    os.makedirs(sparse_model_path, exist_ok=True)

    options = pycolmap.IncrementalPipelineOptions()
    print("Starting incremental mapping...")
    start_time = time.time()
    try:
        reconstructions_dict = pycolmap.incremental_mapping(
            database_path=database_path,
            image_path=image_dir,
            output_path=sparse_model_path,
            options=options
        )
        end_time = time.time()
        print(f"Incremental mapping finished in {end_time - start_time:.2f} seconds.")
        print("reconstuction is:",reconstructions_dict)
        if reconstructions_dict is not None and len(reconstructions_dict) > 0:
            print(f"Mapping generated {len(reconstructions_dict)} reconstruction(s).")

            best_rec_key = -1; max_images = -1
            for rec_id, rec in reconstructions_dict.items():
                 if len(rec.images) > max_images:
                      max_images = len(rec.images); best_rec_key = rec_id

            if best_rec_key != -1 and max_images > 0:
                reconstruction = reconstructions_dict[best_rec_key]
                print(f"Successfully reconstructed {len(reconstruction.images)} images in reconstruction ID {best_rec_key}.")
                print("Image items:",reconstruction.images)
                for k, m in reconstruction.images.items():
                    print(k,m)
                cluster_poses = {}
                successful_pose_extractions = 0
                
                # iterate only on the registered images
                poses = {}
                for img_id in reconstruction.reg_image_ids():
                    img = reconstruction.images[img_id]
                    pose       = get_pose_from_image(img)
                    if pose is None:             
                        continue
                    R, t       = pose
                    poses[img.name] = dict(R=R, t=t)
                    successful_pose_extractions += 1

                if successful_pose_extractions == 0 and len(reconstruction.images) > 0:
                     print("Warning: Failed to extract poses for any registered images.")

                print(f"Finished pose extraction loop for cluster.") # Added for clarity
                print(f"Saving sparse model for reconstruction ID {best_rec_key} to: {sparse_model_path}")
                reconstruction.write(sparse_model_path)
                return poses if successful_pose_extractions > 0 else None

            else:
                print("SfM reconstruction completed but failed to register any images.")
                return None
        else:
            print("SfM reconstruction failed or produced an empty result dictionary.")
            if not os.path.exists(os.path.join(sparse_model_path, 'images.bin')):
                 print("  No sparse model files found.")
            return None

    except Exception as e:
        end_time = time.time()
        print(f"ERROR during SfM reconstruction or pose extraction after {end_time - start_time:.2f} seconds:")
        traceback.print_exc()
        return None
```

```python

all_successful_poses = {}

print("Starting SfM processing for each cluster...")

for cluster_idx, current_cluster_list in enumerate(final_clusters):
    print(f"\n>>> Processing Cluster {cluster_idx} ({len(current_cluster_list)} images) <<<")

    db_prep_result = prepare_colmap_input_for_cluster(
        current_cluster_list,
        cluster_idx,
        WORKSPACE_DIR,
        IMAGE_DIR,
        all_features,
        all_verified_matches
    )

    if db_prep_result:
        database_path, cluster_workspace = db_prep_result

        poses = run_sfm_and_extract_poses(
            database_path,
            IMAGE_DIR,
            cluster_workspace
        )

        if poses:
            print(f"Successfully obtained {len(poses)} poses for Cluster {cluster_idx}.")
            all_successful_poses[cluster_idx] = poses
            for cluster_idx, poses in all_successful_poses.items():
                print(f'\nCluster {cluster_idx}')
                for img_name, P in poses.items():
                    R = P['R']
                    t = P['t']
                    print(f'Image : {img_name}')
                    print('R =')
                    print(R)
                    print('t =', t, '\n')


        else:
            print(f"SfM failed or yielded no poses for Cluster {cluster_idx}.")
    else:
        print(f"Skipping SfM for Cluster {cluster_idx} due to database preparation failure.")

print("Finished processing all clusters.")
print(f"Total number of successfully reconstructed clusters: {len(all_successful_poses)}")
```
